```python
%matplotlib inline
import csv
import sys
import numpy as np
import math
import matplotlib.pyplot as plt
import pylab as pl
import random
```

```python
f = open('data/lbl-pkt-4.csv', 'r')
f.readline()
data = np.loadtxt(f)
```

```python
N = int(math.floor(len(data)/10))
pp = 100
pt = 1000
```

```python
inx = np.zeros(pp)
for i in range(0,pp):
    inx[i] = i
    
PN  = np.zeros(pp)
for i in range(0,pp):
    PN[i] = pt * i

inxN = np.zeros(N);
for i in range(1,N):
    inxN[i] = i

PNN  = np.zeros(N);
for i in range(0,N):
    PNN[i] = pt * i; 

arrCount = np.zeros(pp)
j = 1
for i in range(1,pp):
    while(data[j] < PN[i]):
        arrCount[i] =  arrCount[i] + 1
        j = j + 1

interArr = np.zeros(N)
for i in range(0,N-1):
    interArr[i] = data[i+1] - data[i]

Xshifted = np.zeros((N,pp),int)
for i in range(0,N):
    for j in range(0,pp):
        if (i+j)-1 < N:
            Xshifted[i][j] = data[(i+j)-1]

interArrShifted = np.zeros((N,pp),int)
for i in range(0,N):
    for j in range(0,pp):
        if (i+j)-1 < N:
            interArrShifted[i][j] = interArr[(i+j)-1]

arrAvg = np.zeros(pp)
j = 1
for i in range(0,pp):  
    while(data[j] < PN[i]):
        arrAvg[i] =  arrAvg[i] + data[i]
        j = j + 1
    arrAvg[i] = arrAvg[i] / i
    
interArrAvg = np.zeros(pp)
j = 0
for i in range(0,pp):
    while(data[j] < PN[i]):
        interArrAvg[i] =  interArrAvg[i] + interArr[i]
        j = j + 1
    interArrAvg[i] = interArrAvg[i] / i
```

```python
#Stationarity

inxN = np.zeros(N);
for i in range(1,N):
    inxN[i] = i;

plt.ylabel('Inter-arrival time')
plt.xlabel('Periods')
plt.title('Stationarity')

plt.plot(inx,interArrAvg)
```

```python
# INTENSITY

plt.plot(PN,arrCount)
plt.ylabel('Count of arrivals')
plt.xlabel('Periods')
plt.title('Intensity')
```

```python
countT = np.sum(interArrAvg[1:])
pdfArr = np.zeros(pp)
for i in range(0,pp):
    pdfArr[i] = interArrAvg[i]/countT

plt.ylabel('PDF')
plt.xlabel('Inter-arrivals')
plt.title('PDF of arrival times')   
plt.plot(pdfArr)
```

```python
#IDC

IDC = np.zeros(pp);
l = math.floor(len(interArr)/pp);
for i in range(1,pp):
    IDC[i]  = np.var(interArr[((i-1) * l +1) : (i*l)]) / np.mean(interArr[((i-1) * l +1) : (i*l)] * 1)
plt.xlabel('Time');
plt.ylabel('IDC');
plt.title('IDC of arrivals');
plt.plot(IDC)
```

```python
# Correlation
ccorrInt = np.corrcoef(interArrShifted[0:pp])
```

```python
#IDI

E = np.mean(interArr)
IDI = np.zeros(pp)
for i in range(1,pp):
    IDI[i] = np.var(np.transpose(sum(np.transpose(interArrShifted[0:i])))) /(i * E * E)
plt.plot(PN,IDI)
plt.xlabel('lag')
plt.ylabel('IDI')
plt.title('BC - IDI of inter-arrival times')
```
